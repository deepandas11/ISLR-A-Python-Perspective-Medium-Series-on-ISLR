# Lab Session

```python
df = pd.read_csv('../Data/Boston.csv', index_col=0)
df.head()
```

Let's fit a simple linear model(OLS = Ordinary Least Squares method) with medv as the response and lstat as the predictor

```python
lm = sm.OLS.from_formula('medv ~ lstat', df)
result = lm.fit()

# Printing the Summary
result.summary()

#Printing specified result values
result.rsquared, result.fvalue, result.params.Intercept, result.params.lstat

#Plotting the Regression Line
sns.regplot('lstat','medv', df, line_kws={"color":"r"})
```

```python
new_data = pd.DataFrame([[1,5], [1,6], [1,38]], columns = ['Intercept', 'lstat'])
result.predict(new_data)
```

```python
result.summary()
```

```python
fitted_values = pd.Series(result.fittedvalues, name = "Fitted Values")
residuals = pd.Series(result.resid_pearson, name = "Residuals")
sns.regplot(fitted_values, residuals, fit_reg=False)
```

Multiple Linear Regression

```python
model = sm.OLS.from_formula('medv ~ lstat + age', df)
result = model.fit()
result.summary()
```

```python
# All columns (except medv, which is our response)
model = sm.OLS.from_formula('medv ~ ' + '+'.join(df.columns.difference(['medv'])), df)
result = model.fit()
result.summary()
```

Interaction Terms

```python
model =  sm.OLS.from_formula('medv ~ lstat*age', df)
result = model.fit()
result.summary()
```

Non-Linear

```python
lm.fit2 = sm.OLS.from_formula('medv ~ lstat + np.square(lstat)', df).fit()
lm.fit2.summary()
```
